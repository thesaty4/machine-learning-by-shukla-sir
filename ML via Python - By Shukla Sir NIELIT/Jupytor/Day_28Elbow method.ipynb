```python
#Customer Segmentation Problem
#It is a  Clustering Problem
#It is a Unsupervised Learning bcoz there is no target/outcome column/variable.

import pandas as pd
df=pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
# There are four features in this data set: 
#Gender, Age, Annual Income, Spending Score
# Each entity in this data set represents one customer

#On the basis of which features, you want to segment data set
#Possible answers are: all features or a subset of features

#For this example to begin with, I am using two features
#That is Spending Score, Annual Income

X=df.iloc[:,[3,4]].values
X.shape


```

```python
print(X)
```

```python
#Number of Clusters: Number of groups in which you want your dataset 
#to be divided
```

```python
#Implementation of clustering algorithm K-Means
from sklearn.cluster import KMeans

km=KMeans(n_clusters=5)
#From here also it is clear that it is unsupervised learning bcoz
#in supervised learning fit function takes 2 arguments: X_train,y_train
#but here fit function takes just one argument: X
km.fit(X)

#Fit function creates logic to divide the entire dataset
#by looking at the data
y_predict=km.predict(X)
#Predict function applies that logic to the datasets and put each entity
#of datasets into one of the 5 clusters
print(y_predict)
```

```python
df['Class']=y_predict
```

```python
df.head()
```

```python
df.to_csv("Customer5.csv")
```

```python
#Big Question: How many numbers of clusters are best for my datasets
#Answer: Look at the clusters, visualize it and decide it
#according to your business requirement
```

```python
#is there any method to find optimal number of clusters?
#Yes, there is a method
#Elbow Method is a method to find optimal number of clusters

from sklearn.cluster import KMeans
#Number of clusters : 2 to 10, we are taking
E=[]
for i in range(2,11):
    km=KMeans(n_clusters=i)
    km.fit(X)
    E.append(km.inertia_)
print(E)
    
```

```python
import matplotlib.pyplot as plt
plt.plot(range(2,11),E)
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
#Where Elbow is ending, that is optimal number of clusters 
#for our case this number is 5: this is optimal
#but even if you  take number of clusters as 3 or 4, that is also good
#because it is forming an Elbow
#This method of finding optimal number of clusters is called Elbow 
#Method
```
